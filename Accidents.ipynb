```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data from the CSV file
data = pd.read_csv("accidents.csv")
```

```python
# Count the number of accidents in each neighborhood
neighborhood_counts = data["Neighborhood Name"].value_counts().reset_index()
neighborhood_counts.columns = ["Neighborhood Name", "Number of Accidents"]

# Sort the neighborhoods by the number of accidents in descending order
neighborhood_counts = neighborhood_counts.sort_values(by="Number of Accidents", ascending=False)

#Sort alphabetically
neighborhood_counts_alphabetical = neighborhood_counts.sort_values(by="Neighborhood Name", ascending=True)

#Accidents per neighborhood - will be used in model
print(neighborhood_counts)
neighborhood_counts_alphabetical.to_csv('accidents_extracted.csv')

# Select the top 10 neighborhoods
top_10_neighborhoods = neighborhood_counts.head(10)
```

```python
# Create a bar chart
plt.figure(figsize=(12, 6))
plt.barh(top_10_neighborhoods["Neighborhood Name"], top_10_neighborhoods["Number of Accidents"], color='skyblue')
plt.xlabel("Neighborhood Name")
plt.ylabel("Number of Accidents")
plt.title("Top 10 Neighborhoods with the Most Accidents")
plt.xticks(rotation=45, ha="right")
plt.tight_layout()

# Show the chart
plt.show()
# Save the graph as a PNG file
plt.savefig('Accidents.png', dpi=300, bbox_inches='tight')
```
